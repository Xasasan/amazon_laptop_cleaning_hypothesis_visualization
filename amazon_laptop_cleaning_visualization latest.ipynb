```python
#inches=centimeters÷2.54
def screen_size_clean(value):
    result = str(value)
    result = result.replace('no screen_size', '1')
    if 'Centimetres' in result:
        number = result.split()[0]
        inch = float(number) / 2.54
        return inch
    elif 'Inches' in result:
        number = float(result.split()[0])
        return number
        



df['screen_size(inches)'] = df['screen_size(inches)'].apply(screen_size_clean)
avg = df['screen_size(inches)'].mean()
df['screen_size(inches)'] = df['screen_size(inches)'].fillna(avg)


    
```

```python
def price_clean(value):
    result = str(value)
    result = result.lower().strip()
    if 'no price' in result or result == 'nan' or result == 'none' or result == '':
        return np.nan
    if 'no price' in result:
        return np.nan
    if ',' in result:
        result = result.replace(',', '')
    return float(result)




df['price(₹)'] = df['price(₹)'].apply(price_clean)
avg_price = df['price(₹)'].mean()
print(avg_price)
df['price(₹)'] = df['price(₹)'].fillna(avg_price)
nan_count = df['price(₹)'].isnull().sum()
print(f"Missing values BEFORE fillna: {nan_count}")
```

```python
#GB = MB / 1024   # binary system (most common in computing)
def ram_clean(value):
    result = str(value)
    result = result.replace('no ram', '1')
    if 'GB' in result:
         mb = result.split()[0]
         return float(mb)
    elif 'MB' in result:
         mb = float(result.split()[0])
         mb = mb / 1024
         return mb
        

        
   

df['ram'] = df['ram'].apply(ram_clean)
avg_price = df['ram'].mean()
print(avg_price)
df['ram'] = df['ram'].replace('1', avg_price)
df['ram'] = df['ram'].fillna(avg_price)



    
```

```python
#GB = MB / 1024   # binary system (most common in computing)
def storage_clean(value):
    result = str(value)
    result = result.replace('no storage', '1')
    if 'GB' in result:
        mb = result.split()[0]
        return int(float(mb))
    elif 'MB' in result:
        mb = float(result.split()[0])
        mb = mb / 1024
        return mb
    elif 'TB' in result:
        mb = float(result.split()[0])
        mb = int(mb) * 1024
        return int(mb)
         
 
   

df['storage'] = df['storage'].apply(storage_clean)
avg_price = df['storage'].mean()
print(avg_price)
df['storage'] = df['storage'].replace('1', avg_price)
df['storage'] = df['storage'].fillna(avg_price)
    
```

```python
# 8-gipoteza: Ryzen 7 vs Core i5 price comparison (vertical bar)

amd = df[df['cpu_model'] == 'Ryzen 7']['price(₹)'].mean()
intel = df[df['cpu_model'] == 'Core i5']['price(₹)'].mean()

print(f"Ryzen 7 average price: {amd}")
print(f"Core i5 average price: {intel}")

models = ['Ryzen 7', 'Core i5']
avg_prices = [amd, intel]

plt.figure(figsize=(7, 10), dpi=120)
plt.bar(models, avg_prices, color=['orange', 'skyblue'], edgecolor='black')
plt.ylabel('Average Price (₹)')
plt.xlabel('CPU Model')
plt.title('Average Laptop Prices: Ryzen 7 vs Core i5', 
          fontdict={'fontname': 'Times New Roman', 'fontsize': 18})
plt.show()
```
